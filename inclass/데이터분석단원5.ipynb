# mpg 통합 연비 변수 만들기

```python
mpg = pd.read_csv('data/mpg.csv')
mpg.sample(8) # sample 메소드를 이용하면 램덤으로 알려주며 (8)이라 했기떄문에 8개까지만 보임
```

```python
# 통합 연비 변수(total) 만들기, total <= cty + hmy의 평균: 시험 예상
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

```python
mpg[['total']]
```

```python
#통합 연비 변수(mpg_mean) 만들기, total <= cty + hwy으; 평균: 시험 예상
mpg['mpg_mean'] = mpg[['cty', 'hwy']].mean(axis='columns')
mpg.head()
```

```python
mpg[['mpg_mean']]
```

```python
sum(mpg['total']) / len(mpg) # total 합계를 행 수로 나누기
```

```python
mpg['total'].mean() # 통합 연비 변수 평균
                    # df_mean 메소드로 평균을 간단하게 구할 수 있다.
```

## 중첩 조건문 활용하기

```python
mpg
```

```python
# total 기준으로 A,B,C 등급 부여
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
               np.where(mpg['total'] >= 20, 'B', 'C'))

mpg.head()
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
# 순서대로 하기위해서 sort_index() 사용
```

```python
count_grade.plot.bar(rot = 0);
```

```python
import seaborn as sns

# 'grade' 열에 대한 빈도수를 퍼센트로 표시하는 countplot 그리기
sns.countplot(data=mpg, x="grade", hue="grade", stat='percent');
```

# 정리하기

```python
#1. 패키지 로드
import pandas as pd
import numpy as np

# 2. 데이터 불러오기
mpg = pd.read_csv('mpg.csv')

#3. 데이터 파악하기
mpg.shape          # 행, 열, 수
mpg.head()         # 데이터 앞부분
mpg.tail()         # 데이터 뒷부분
mpg.info()         # 속성
mpg.describe()     # 요약 통계량

# 4. 변수명 바꾸기
mpg = mpg.rename(columns = {'manufactuer' : 'company'})

# 5. 파생변수 만들기
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2    # 변수 조합
mpg['test'] = np.where(mpg['total']>=20, 'pass', 'fail')   # 조건문 활용

# 6. 빈도 확인하기
count_test = mpg['test'].value_counts()
count_test.plot.bar(rot = 0)
```
