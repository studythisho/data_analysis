# 단원4 데이터프레임의 세계

````python
import pandas as pd

df = pd.DataFrame({'name' : ['김지훈', '이유진', '박동현', '김진지'],
                   'english' : [90, 80, 60, 70],
                   'math' : [50, 60, 100, 20]})
df
````

````python
df.info() # info는 간단하게 알려줌
````

````python
df['english']
````

````python
sum(df['english'])
````

````python
sum(df['math'])
````

````python
sum(df['english']) /4 # = len(df)
````

````python
df.describe() # 간단한 통계적 요약 정보를 제공하는 함수입니다.
              # 이 메소드를 호출하면 데이터의 중심 경향성(평균, 중앙값)
              # 분포의 퍼짐 정도(표준편차, 사분위 범위)
              # 및 분포의 형태(최소값, 최대값, 사분위수)
              # 에 대한 요약 통계를 제공합니다.

````

````python
df.describe(include = 'all')
````

````python
df.sum()
````

````python
df.sum(numeric_only=True) # 숫자형 열에 대해서만 요약 정보를 생성
````

````python
df.sum(axis=0, numeric_only=True) # 연산이 행(row) 방향으로 수행
````

````python
df.sum(axis="index", numeric_only=True)
````

````python
df.mean(axis="index", numeric_only=True)
````

````python
df.sum(axis=1, numeric_only=True) # 연산이 열(column) 방향으로 수행
````

````python
df.sum(axis="columns", numeric_only=True)
````

````python
df.mean(axis="columns", numeric_only=True)
````
