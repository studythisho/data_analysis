```python
df = sns.load_dataset('titanic')
df
```

```python
sns.countplot(data = df, x = 'class') # 선실 등급
```

```python
sns.countplot(data = df, x = 'class', hue = 'alive') # hue는 변수의 항목별로 막대의 색을 다르게 표현 파라미터
```

```python

sns.countplot(data = df, y = 'class', hue = 'alive')
```

```python
df.info()
```

```python
sns.countplot(data=df, x='sex')
```

```python
df.pclass.unique() # 타이타닉에 대한 class에 p를 붙이면 총 열로 나타내줌
                  # 여기서는 unique()의 메소드르 썼기에 배열로 알려줌
```

```python
df["pclass"].unique() # 다르게 하는 또다른 코드
```

```python
df.pclass.values
```

```python
df.pclass.value_counts() # class 등급별 총 인원수를 알려줌
```

```python
df.nunique()
```

```python
df[['pclass', 'who']].values
```

```python
df.groupby('sex').count() # 데이터프레임(df)에서 'sex' 열을 기준으로 그룹화(groupby)하고
                          # 각 그룹 내의 레코드 수(count)를 계산하는 작업을 수행
```

```python
df.groupby('class').count()
```

```python
df.groupby('alive').count()
```
